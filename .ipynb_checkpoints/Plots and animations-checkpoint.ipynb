# Plots and animations in Python


## Basic plotting
The simplest way to plot a curve in Python is to use the `matplotlib` package, designed to emulate Matlab style plot commands. 

There are extensive tutorials over at [matplotlib.org](https://matplotlib.org/stable/tutorials/index) so this notebook will have just a few examples to get yourself familiar with the absolute basics.

Here's a basic example. We import `numpy` which is our general purpose numerical computing package.

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(0, 1)
y = np.sin(2*np.pi*x)

plt.plot(x,y, 'k')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.grid()

plt.show() # use this line to prevent annoying text from showing (though it is required if running a standalone python script)
```

Here's an example with multiple subplots

```python
x = np.linspace(0, 1)
y = np.sin(2*np.pi*x)
z = np.sin(4*np.pi*x)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2,  # decide number of columns and rows
                                             tight_layout=True,  # makes sure the axes and labels fit nicely in the figure
                                             sharex='col', sharey='row') # x axes are shared for each column, and y axes are shared for each row 


ax1.plot(x,y, 'k')
ax1.grid()
ax1.set_ylabel('Amplitude')

ax2.plot(x,y**2, 'k')
ax2.grid()

ax3.plot(x,y*z, 'k')
ax3.grid()
ax3.set_xlabel('x')
ax3.set_ylabel('Amplitude')

ax4.plot(x,z**2, 'k')
ax4.grid()
ax4.set_xlabel('x')


plt.show() # use this line to prevent annoying text from showing (though it is required if running a standalone python script)
```

## Animations 

`matplotlib.animation` can be used to display and export animations. You have to create a function (here named `update()`) which updates the state of the function or data you want to show for each frame of the animation.  

```python
# required imports

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

plt.rcParams["animation.html"] = "jshtml"

duration = 3 # seconds
fps = 30
frames = duration * fps
dt = 1/fps

# initialise input data 
x = np.linspace(0,1)
y = np.sin(2*np.pi*x)

# create plot(s)

fig, ax = plt.subplots()
omega = 2*np.pi
line, = ax.plot(x, y, 'k')
ax.grid()

def update(frame):
    y = np.sin(2*np.pi*x + omega*dt*frame)
    line.set_ydata(y)
    return line,

plt.close()
# Create animation
animation.FuncAnimation(fig, update, frames=frames, interval= int(1000 * dt))
```
