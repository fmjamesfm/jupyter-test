## The wave equation

We looked at the heat equation in [quite some detail](Fourier.ipynb):

$$
\frac{\partial u}{\partial t} = D \frac{\partial^2 u}{\partial x^2}
$$
Among other things, we found that time dependence $T(t)$ was a steady decay to equilibrium, $T=e^{-\lambda t}$. Energy is dissipated over time. 

Now consider the wave equation, which, at a glance looks quite similar:

$$
\frac{\partial^2 u}{\partial t^2} = c^2 \frac{\partial^2 u}{\partial x^2} \tag{1}
$$

We track a quantity $u$, which might represent the displacement of a string under tension from equilibrium. $c$ is the speed of wave propagation (or speed of sound). The difference between this and the heat equation is that the time derivative is of second order. This seemingly small difference has profound implications, as it changes the time domain dynamics from steady decay and loss of energy, to oscillation and conservation of energy. 

We can already say a few things without going into solutions. 

It's helpful to think of a bi-directional cause and effect relationship between both sides of the equation:

* The right hand side,  $\frac{\partial^2 u}{\partial x^2}$,  represents curvature: how sharply $u$ bends.
* The left hand side, $\frac{\partial^2 u}{\partial t^2}$, is acceleration: how quickly $u$ speeds up or slows down

The local curvature of $u$ drives changes in speed (acceleration), which in turn changes the local curvature.

Suppose the string is initially curved upwards at point $A$. Then $A$ is lower than its neighbours, and the curvature is positive. This creates a net upward acceleration: $A$ gains speed, overshoots its neighbours, and the curvature flips. The sign of acceleration reverses, slowing $A$ down and pulling it back. The result is oscillation. The feedback between curvature and acceleration underlies all wave behavior. 

The position at each point affects its neighbours, so if the string is stationary, and there is a change in displacement at some point $A$ (for example due to an external force), this change will transmit to its neighboring points (since they will be experiencing increasing curvature) and they will in turn transmit to their neighbours and so on. Any local disturbance travels away in all directions at a speed of $c$, and so we have wave propagation. 

**Disturbances in space are smoothed out by the heat equation, while the wave equation transmits them!**

## Discretising the wave equation: the finite difference approach

We will jump straight into numerical implementation of a particular case: a string of length $L$ under tension and rigidly fixed at both ends. 
The string has mass per unit length $\mu$ and tension $T$. The ends being fixed means displacement $u$ is constant there. We will say that $u=0$ at $x=0$ and $x=L$. We're not going to analytically solve this for now, it will be much easier to just set $u$ to zero in our code where appropriate.

We'll first need to discretise the domain, so we divide the string into a finite number of points $x_i$, spaced $\Delta x$ apart, and define an array to keep track of the displacement $u$:

$$
u(x_i) = \{u(x_0), u(x_1), u(x_2), ... \}
$$

We decide a time step $\Delta t$, which should ideally be small enough not to cause numerical trouble:

$$\Delta t < \frac{\Delta x}{c}$$ 

The curvature of the string $\frac{\partial^2}{\partial x^2}u(x_i)$ will be approximated using a central difference scheme:

$$
\frac{\partial^2 u(x_i)}{\partial x^2} \approx \frac{u_i(x_{i+1})-2u_i(x_i)+u_i(x_{i-1})}{\Delta x^2}
$$

So that, for each time step, the acceleration $\ddot{u}$ (using double dots to indicate double differentiation with respect to time) at a position $x_i$ will be given by the approximate curvature times the speed of sound squared:

$$
\ddot{u}_i \approx c^2 \frac{u(x_{i+1})-2u(x_i)+u(x_{i-1})}{\Delta x^2}
$$

We then set the velocity from the acceleration $\dot{u}$:

$$\dot{u_i} = \dot{u}_{i-1} + \Delta t \cdot \ddot{u}_i$$

And finally the position $u$ is updated using its rate of change:

$$u_{i+1}= u_{i} + \Delta t \cdot \dot{u_{i}}$$

We will discretise a string into N spatial points, decide a suitable time interval, and some initial conditions so that we can play out what happens over time. You will see that a local disturbance travels away in all directions. Along a one dimensional string there are only two possible directions: left and right.


```python
# required imports
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import scipy.signal 
plt.rcParams["animation.html"] = "jshtml"


N = 500 # number of discrete points along the string 
L = 1

x = np.linspace(0,L, N) # x_i coordinates
dx = x[1]-x[0] # spatial step


# initial conditions
width = 50 # factor for the width of the initial displacement
u0 = 0.5*np.exp(-0.5*((x-L/2)*width)**2) # initial displacement

plt.plot(u0)

du0 = np.zeros(u0.shape) # initial velocity (du=0, the string starts stationary)


c = 1 # speed of wave propagation (distance per unit time)
dt = 0.9 * dx / c # time step (small enough not to cause problems, but you might want to see what happens if it's too big...)
r = 0.1 # damping coefficient: increase this to see what happens


# heat equation constants

D = 1 # diffusion coefficient

# animation setup 
duration = 3 
fps = 30 # frames per second
steps_per_frame = int((1/dt)/fps) # steps per second
frames = fps * duration 

def iterate_wave(u, du):
    ddx = np.zeros_like(u)
    ddx[1:-1] = (u[0:-2] + u[2:] - 2*u[1:-1]) / dx**2
    ddu = c**2* ddx - du*r # this is the acceleration. The term du*r represents damping (a force in the opposite direction to the local velocity du)  
    du_next = du + dt*(c**2*ddx - du*r)
    u_next = u + dt*du_next
    return u_next, du_next
    
def iterate_heat(u, du):
    ddx = np.zeros_like(u)
    ddx[1:-1] = (u[0:-2] + u[2:] - 2*u[1:-1]) / dx**2
    
    du_next = du + dt*(- D*du)
    u_next = u + dt*du_next
    
    return u_next, du_next

state1 = [u0.copy(), du0.copy()]
state2 = [u0.copy(), du0.copy()]

def update(frame):
    if frame==0:
        line.set_ydata(state1[0])
        line.set_ydata(state2[0])
        return line1, line2
        
    for i in range(steps_per_frame):
        # multiple simulation steps per animation frame    
        state1[0], state1[1] = iterate_wave(state1[0], state1[1])
        state2[0], state2[1] = iterate_heat(state2[0], state2[1])
        
    line1.set_ydata(state1[0])
    line2.set_ydata(state2[0])
        
    return line1, line2
    
# Set up figure
fig, (ax1, ax2) = plt.subplots(nrows=2)
ax1.set_ylim([-0.5,0.5])
ax2.set_ylim([-0.5,0.5])

line1, = ax.plot(x, u0, 'k')
line2, = ax.plot(x, u0, 'k')

ax.set_axis_off()
plt.close()
# Create animation
animation.FuncAnimation(fig, update, frames=frames, interval=20)
```

We introduced some disturbance (with the initial displacement `u0`), and solved the wave equation step by step from time zero. We can see the disturbance travels out from the centre in two separate "wavelets" that in turn reflect from the fixed edges of the string. The reflections flip the sign of the travelling wavelets.
    
We could instead decide that the string start stationary, and introduce an external force at some point. In the example code below, we will add the forcing function $f(t)$

$$
f(t)= \cos(\omega t)
$$

This force is introduced as an additional term in the wave equation at some point $x_f$:

$$
\frac{\partial^2 u}{\partial t^2} = c^2 \frac{\partial^2 u}{\partial x^2}+\delta(x_i-x_f) f(t)
$$

Here we're using the Dirac delta $\delta(x_i - x_f)$ to show that the forcing term only appears at position $x_f$. The rest of the string is free to move according to the unaltered wave equation.

For the forcing function, we're picking a value of $\omega$ to correspond to one of the natural frequencies of the string fixed at both ends. These are a function of length $L$:

$$
\omega_n = \frac{n \pi}{L} 
$$
where $n \in \{1,2,3,...\}$.
Have a go at the code below; you could for example try changing the frequency or position of the external force.

Exciting a single frequency is showing us something quite interesting: after a short time the string starts moving in a *standing wave* pattern.

## D'Alembert's solution for the one dimentional problem
Let's jump back to the one dimensional wave equation:

$$
\frac{\partial^2 u}{\partial t^2} = c^2 \frac{\partial^2 u}{\partial x^2} \tag{1}
$$

If we pick two arbitrary constants $a$ and $b$, the derivative of a function of the type $f(ax+bt)$ with respect to $x$ can be written, by the chain rule:

$$
\frac{\partial}{\partial x} f(ax+bt) = af'(ax+bt)
$$

And the second derivative would be:

$$
\frac{\partial^2}{\partial x^2} f(ax+bt) = a^2f''(ax+bt)
$$

With respect to $t$ the second derivative would be: 

$$
\frac{\partial^2}{\partial t^2} f(ax+by) = b^2f''(ax+bt)
$$

Let's consider now a function $u(x,t)=f(x+ct)$. Its second order partial derivatives are:

$$
\frac{\partial^2}{\partial t^2} u = c^2 f''(x+ct)\text{ ,  } 
\frac{\partial^2}{\partial x^2} u = f''(x+ct) 
$$

So this satisfies the 1D wave equation:

$$
\frac{\partial^2 u}{\partial t^2} = c^2 \frac{\partial^2 u}{\partial x^2} \tag{1}
$$

$u(x,t) = f(x-ct)$ works just as well. And because the wave equation is linear, any linear combination of these functions is also a solution:

$$u(x,t)=F(x-ct)+G(x+ct)$$

The arguments of these functions are quite interesting! Let's pick one of the two solutions and say that $u(x,t)=F(x-ct)$. Now we look at coordinates $(x_0, t_0)$, where the displacement is $u(x_0,t_0)=F(x_0 - c t_0)$.
We can prove that it makes no difference whether I remain at time $t_0$ and look a distance $L$ to the left (in the negative direction), or wait for a time $L/c$ at the same location:

\begin{aligned}
u(x_0-L, t_0) = F((x_0 - L) - ct_0) & = F(x_0-ct_0 - L)  \\
u(x_0, t_0 + L/c) = F(x_0 - c (t_0 - L/c)) & = F(x_0-ct_0-L)  
\end{aligned}

A value on the left will move to the right over time. In other words, the function $F$ as a whole is moving to the right! The function $G$ is the same, except the difference in sign tells us that it is travelling to the left.

Time and space variables here are interchangeable down to a constant $c$! This symmetry between time and space is a fundamental aspect of waves.


## Solution by separation of variables 

Separation of variables here, as before with $u=X(x)\cdot T(t)$ gives us:

$$
\frac{T''}{T}=c^2\frac{X''}{X}=-\omega^2
$$

Here we use a constant appropriately named $\omega^2$. Now we extract, as we have for the heat problem, two equations to solve for $X$ and $T$ independently. This time, the two equations have essentially the same form: a second order ODE. The solutions to these are oscillatory due to the negative sign chosen for the constant on the RHS.

$$
T''=-\omega^2 T
$$

So valid solutions are $T(t)=\sin( \omega t)$, or $T(t)=\cos (\omega t)$ and linear combinations as before. We tend to pick

$$T(t)=e^{i\omega t}$$

We call this harmonic time dependence. It's the key difference between this problem, and the heat equation we've seen, and it tells us that in this case physical quantities like $u$ oscillate over time.



Let's take a step back and consider what conditions lead us to the wave equation in the first place. This depends on the physical system we're examining, but it usually results from a combination of inertia and a restoring force. 

Newton's second law describes the motion of an object of mass $m$ subject to a net external force $F$:

$$F=m a=m \frac{d^2u}{dt^2}$$

If the object is attached to a spring, the spring produces a restoring force that opposes the motion $u$ as long as $u$ is away from equilibrium ($u=0$):

$$F=-ku$$

This is Hooke's law. We can already see that by equating these forces, we end up with the same time dependence as the wave equation:

$$\frac{d^2u}{dt^2}=-\omega^2u$$ 

with a conveniently defined $\omega_0^2 = \frac{k}{m}$. (Compare this with $\frac{d^2 T(t)}{dt^2}=-c^2 T(t)$)

We call it $\omega_0$, because when solving this ODE it has only one value, the only allowed value. It is the so-called *natural frequency* or *eigenvalue* of this system. Why is this so? This equation takes the form of an eigenvalue problem:

$$
\mathcal{L}(u) = \lambda u
$$

Where $\mathcal{L}$ is a linear differential operator (in the case of the mass-spring system it is a second order differentiation with respect to time). 


$\omega$ is termed the *angular frequency* (it has units of radians per unit time) but can also be thought of as angular velocity: how quickly the vector $u=Ae^{i\omega t}$ (which is the the solution) rotates.
