## Two ports


Recall the (complex) general solution to the wave equation in one dimension as a sum of two traveling plane waves:

$$
p(x) = A e^{-ikx} + B e^{ikx}
$$

Recall also that for a bounded domain this could require a series expansion with individual $k_n$, $A_n$ and $B_n$.

We assume here harmonic time dependence - so much so that we will say that $p(x,\omega)=p(x)e^{i\omega t}$ once and henceforth neglect to write the term $e^{i\omega t}$ entirely unless forced by necessity. This is acoustics, of course everything oscillates over time at frequency $\omega$. 

Now we have our general solution, let's impose some conditions. A boundary appears at $x=0$. This will be the interface between two media. On the left (negative direction) of the boundary we have our reference medium, air, with characterstic impedance given by $z_0=p/v=\rho_0 c_0$. We have two travelling waves there of amplitudes $A$ and $B$ travelling in opposite directions. 

We'll need acoustic pressure and velocity via Euler's equation:

$v(x)=-\frac{1}{i\omega \rho} \frac{\partial p(x)}{\partial x}$

On the right hand side we have a medium with given (possibly complex) characteristic impedance $z_1=p/v=\rho_1 c_1$. We now introduce another boundary on the right of this medium at $x=L$. Further right than $L$ we have our reference medium again. Now it looks like we're approaching a sound transmission problem. Incident waves arrive from the left hand side, travel through the medium with $k_1$ and leave to the right. We don't have reflected waves arriving from the right hand side. Here's a diagram of this system:


```python
import numpy as np
import matplotlib.pyplot as plt 

fig, ax = plt.subplots()

omega = 2*np.pi * 200
rho0 = 1.213
c0 = 343

rho1 = 2 
c1 = 300


k0 = omega/c0
k1 = omega/c1
theta0 = np.pi/2

kx1 = np.sqrt(k1**2 - kx0**2)

z0 = rho0 * c0
z1 = rho1 * c1 


L = 1

ax.axvline(0, color='k')
ax.axvline(L, color='k')
ax.set_xlim(-L, 2*L)
ax.set_ylim(-1, 1)

ax.quiver(-L,-0.5,L,0.5, angles='xy', scale=1, scale_units='xy')
ax.quiver(0,0,-L,0.5, angles='xy', scale=1, scale_units='xy')

ax.quiver(-L,-0.5,L,0.5, angles='xy', scale=1, scale_units='xy')
ax.quiver(0,0,-L,0.5, angles='xy', scale=1, scale_units='xy')

plt.show()

```
